```python
class my_dialect(csv.Dialect):
    lineterminator = "\n"
    delimiter = ","
    quotechar = "|"
    quoting = csv.QUOTE_MINIMAL
    
```

```python
f = open('/home/akshat/Desktop/Python_data_analysis/dummy_data/db_quotes.csv')
reader = csv.reader(f,dialect= my_dialect)
```
